# Exploratory Analysis and Visualization

```python
plt.figure(figsize=(14,7))
sns.scatterplot(data=raw_data, x='Age', y='SkinThickness', hue='Outcome')
plt.title('Relationship of Age with skin thickness and outcome');
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=raw_data, x='Outcome')
plt.title('Number of people with and without Diabetes')
plt.ylabel('Number of people')
plt.grid();
```

```python
plt.figure(figsize=(12,5))
plt.pie(x=raw_data.Outcome.value_counts(), labels=raw_data.Outcome.unique(), autopct='%.1f%%')
plt.title('Percent of those with Diabetes and those without');
```

# Questions and Answers

## 1. How many people had a Diabetes Pedigree Function greater than 2.42?

```python
a = raw_data[(raw_data['DiabetesPedigreeFunction'] >= 2.42)]
a
```

```python
a.shape
```

```python
print('''
How many people had a Diabetes Pedigree Function greater than 2.42?
Answer: {}'''.format(a.shape[0]))
```

## 2. How many people had a normal BMI and had Diabetes?

```python
# normal BMI = (18.5 to 24.9)
b = raw_data[(raw_data['BMI'] >= 18.5) & (raw_data['BMI'] <= 24.9) & (raw_data['Outcome'] == 1)]
b
```

```python
b.shape
```

```python
print('''
How many people had a normal BMI and had Diabetes?
Answer: {}'''.format(b.shape[0]))
```

## 3. How many people had a normal Glucose level and has diabetes?

```python
c = raw_data[(raw_data['Glucose'] >=70) & (raw_data['Glucose'] <=100) & (raw_data['Outcome'] == 1)]
c
```

```python
c.shape
```

```python
print('''
How many people had a normal Glucose level and has diabetes?
Answer: {}'''.format(c.shape[0]))
```

## 4. How many people had a normal Glucose level and didn't have diabetes?

```python
d = raw_data[(raw_data['Glucose'] >=70) & (raw_data['Glucose'] <=100) & (raw_data['Outcome'] == 0)]
d
```

```python
d.shape
```

```python
print('''
How many people had a normal Glucose level and didn't have diabetes?
Answer: {} people'''.format(d.shape[0]))
```
